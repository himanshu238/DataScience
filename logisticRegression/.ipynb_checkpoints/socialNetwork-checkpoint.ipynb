```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
data.head()
```

```python
data.drop('User ID',axis=1,inplace=True)
```

```python
data.head()
```

```python
data.Gender.replace('Male',1,inplace=True)
```

```python
data.Gender.replace('Female',0,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
data.drop(['Purchased','Gender'],axis=1)
```

```python
x = sc.fit_transform(data.drop(['Purchased','Gender'],axis=1))
```

```python
x = pd.DataFrame(x,columns=['A','E'])
```

```python
x = pd.concat([x,data['Gender']],axis=1)
```

```python
x
```

```python
purchased_count =0
for i in data.Purchased:
    if i==1:
        purchased_count+=1
print(purchased_count)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm  = SMOTE()
```

```python
x, y = sm.fit_resample(x,data.Purchased)
```

```python
purchased_count =0
for i in y:
    if i==0:
        purchased_count+=1
print(purchased_count)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
regressor = LogisticRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
