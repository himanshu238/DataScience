## Data Preprocessing

```python
data['sex'].replace('female',0,inplace=True)
```

```python
data['sex'].replace('male',1,inplace=True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder() #ordinal Encoding type other is target guided
```

```python
data['smoker'] = le.fit_transform(data['smoker'])
```

```python
data.head()
```

```python
data.keys()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder() #nominal encoding type
```

```python
ohe.fit(data[['region']])
```

```python
ohe.categories_
```

```python
col = ohe.transform(data[['region']]).toarray()
```

```python
col
```

```python
col = pd.DataFrame(col,columns=['northeast', 'northwest', 'southeast', 'southwest'])
```

```python
col.drop('southwest',axis=1,inplace = True)
```

```python
da = pd.concat([data,col],axis=1)
```

```python
da
```

```python
da.drop('region',axis=1,inplace = True)
```

```python
da
```

```python
y = da['charges']
x = da.drop('charges',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=101) #splitting data
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
print(r2_score(y_test,y_pred))
```

```python
mean_squared_error(y_test,y_pred,squared=False)
#rmse
```

```python
mean_squared_error(y_test,y_pred,squared=True)
#mse
```

```python
mean_absolute_error(y_test,y_pred)
```
